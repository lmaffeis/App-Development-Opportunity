## Exploring data

We start by opening and exploring our two dataset. To make them easier to explore, we create a function named `explore_data`. The  `explore_data` function:
* Takes in four parameters:
    * `dataset`, which is expected to be a list of lists.
    * `start` and `end`, which are both expected to be integers and represent the starting and the ending indices of a slice from the data set.
    * `rows_and_columns`, which is expected to be a Boolean and has `False` as a default argument. 
* Slice the data set using `dataset[start:end]`.
* Loops thorough the slice, and for eache iteration pronts a row and adds a new line adter that row using `print('\n')`.

### Duplicates elimination from the Android dataset

If we read the discussion section of the Google data set, it is reported that some apps have duplicate entries. For instance, Instagram has four entries:

```python
for i in android:
    name = i[0]
    if name == 'Instagram':
        print (i)
```

We want to check and isolate all the apps that have duplicates inside the Google data set. In order to do that, we:
- Create two list: one for storing the name of duplicate apps, and one for storing the name of unique apps.
- Loop through the `Android` data set and for each iteration:
    - We save the app name to a variable named `name`
    - If `name` was already in the `unique_apps` list, we append `name` to the `duplicate_apps` list
    - Else, we appended `name` to the `unique_apps` list

Since we don't want to count certain apps more than once, when we analize data, we eliminate the duplicates entry according to the number of reviews of the app (contained in the "Reviews" column). We only keep the row with the highest number of reviews and remove the other entries for any given app.

```python
#Android data set
duplicate_apps = []
unique_apps = []

for apps in android:
    name = apps[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)
        
print ('Number of duplicated apps: ', len(duplicate_apps))
print ('Number of unique apps: ', len(unique_apps))
```

```python
#Apple data set
duplicate_apps = []
unique_apps = []

for apps in ios:
    name = apps[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)
        
print ('Number of duplicated apps: ', len(duplicate_apps))
print ('Number of unique apps: ', len(unique_apps))
```

We see that in the Google Play data set, 1,181 observations are duplicates. After we remover this duplicates we should be left with 9,659 rows.
To remove these duplicates, we will:
- Create a dictionary, where each dictionary key is a unique app name and the corresponding dictionary value is the highest number of reviews of the app
- Use the information stored in the dictionary and create a new dataset, which will have only one entry per app 

Let's start by creating the dictionary:

```python
# Creation of the dictionary:
reviews_max = {}
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    if name in reviews_max and reviews_max[name]<n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
```

We expect the lenght of the dicitionary equal to 9,659, since it only contains unique values



```python
len(reviews_max)
```

We use the dictionary we created to remove duplicates rows. In order to do that:
* We start by creating two lists: `android clean`, that stores the clean dataset and the `already_added` which will just store app names
* We loop through the Android data set and for each iteration:
    * Assign the app name to a variable named `name`
    * Convert the number of reviews to float, and assign it to a vairable named `n_reviews`.
* If `n_reviews` == number of reviews contained of the app `name` (contained in the dictionary), and name is not in `already_added`:
    * We append the entire row to the `android_clean` list
    * Append the name of the app `name` to the `already_added` list

```python
android_clean = []
already_added = []
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```

### Removing non-Eglish apps

We use English for the apps we want to develop at our company and we would like to analyze only the apps that are directed toward an English-speaking audience. If we explore the data long enough, we will find that both data sets ahve apps with names that suggest thay are not directed toward an Enlgish-speaking audience.
We report below a couple of examples for each data set:

```python
print (ios[813] [1])
print (ios [6731] [1])
print (android_clean [4412][0])
print (android_clean[7940][0])
```

We are not interested keeping these apps, so we are going to remove them.
One way to go about this is to remove each app with a name containing a symbol that is not commonly used in English text — English text usually includes letters from the English alphabet, numbers composed of digits from 0 to 9, punctuation marks (., !, ?, ;), and other symbols (+, *, /).

Behind the scenes, each character we use in a string has a corresponding number associated with it. For instance, the corresponding number for character `a` is 97, character `A` is 65, and character `爱` is 29,233. We can get the corresponding number of each character using `ord()`

- We first write a function that takes in a string and return `False` if there's any character in the string that doesn't belong to the set of common English characters, otherwise it returns `True`. 
- Inside the function, we iterate over the input string. For each iteration we check whether the number associated with the character is greater than 127. When a character is greater than 127, the function should immediately return `False` — the app name is probably non-English since it contains a character that doesn't belong to the set of common English characters.
- If the loop finishes running without the return statement being executed, then it means no character had a corresponding number over 127 — the app name is probably English, so the functions should return `True`

```python
def is_english(string):
    for letter in string:
        if ord(letter) >127:
            return False
    return True
```

We check the correctness of our function:

```python
print(is_english('Instagram'))
print(is_english('爱奇艺PPS -《欢乐颂2》电视剧热播'))
```

However if we pass the string 'Instachat 😜' ,over function recognizes the app as non-english: this is because emojis fall outside the ASCII range and have corresponding values over 127

```python
print(is_english('Instachat 😜'))
```

We need to modify our previous function, in order to avoid to lose useful data for our analysis. To minimize the impact of data loss, we'll only remove an app if its name has more than three characters with corresponding numbers falling outside the ASCII range. This means all English apps with up to three emoji or other special characters will still be labeled as English. Our filter function is still not perfect, but it should be fairly effective.

```python
def is_english(string):
    counter = 0
    for word in string:
        if ord(word) > 127:
            counter +=1
    if counter >3:
        return False
    return True
```

As we can see below, the function now works correctly:

```python
print(is_english('Instachat 😜'))
print (is_english('爱奇艺PPS -《欢乐颂2》电视剧热播'))
```

We use the `is_english` function to filter out non-english apps from both datasets. We loop through each dataset and if the app is identified as english we append the whole row to a separate list

```python
#Filter for the Android and iOS data set:
android_english = []
ios_english = []

for app in android_clean:
    name = app[0]
    if is_english(name) == True:
        android_english.append(app)

for app in ios:
    name = app[1]
    if is_english(name) == True:
        ios_english.append(app)
        
explore_data(android_english, 0, 3, True)
print('\n')
explore_data(ios_english, 0, 3, True)
```

## Most common Apps by genere

As we mentioned in the introduction, our aim is to determine the kinds of apps that are likely to attract more users because our revenue is highly influenced by the number of people using our apps.

Let's begin the analysis by getting a sense of what are the most common genres for each market. For this, we'll need to build frequency tables for a few columns in our data sets.

The columns we are going to use to find out what are the most common genres in each markets are:
 -  'Prime_genre' for the iOS data set
 - 'Genres' and 'Category' for the Android data set

We build **two** functions to analyze the frequency tables:
1. One function to generate frequency tables that show percentage
2. Another function we can use to display the percentage in a descending order

We are able to build dictionaries to build frequencty tables, however dictionaries do not have orders, therefore we need to build the function n.2 described in the pargraph above.

- First of all we are going to build a `freq_table()` function, that comput the observed frequency of the apps sorted by genres.
- We then define a `display_table` function that:
    1. Takes in two parameters: `dataset` and `index`. `dataset` is expected to be a list of lists, and `index` is expected to be an integer.
    2. Generates a frequency table using the `freq_table()` function.
    3. Transforms the frequency table into a list of tuples, then sorts the list in a descending order.
    4. Prints the entries of the frequency table in descending order.

```python

def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage 
    
    return table_percentages


def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

As a workaround, we'll take the total number of user ratings as a proxy, which we can find in the rating_count_tot app.

We calculate the average number of user ratings per app genre on the App Store

```python
genres_ios = freq_table(ios_final, -5)

for genre in genres_ios:
    total = 0
    len_genre = 0
    for app in ios_final:
        genre_app = app[-5]
        if genre_app == genre:            
            n_ratings = float(app[5])
            total += n_ratings
            len_genre += 1
    avg_n_ratings = total / len_genre
    print(genre, ':', avg_n_ratings)
```

We're going to leave the numbers as they are, which means that we'll consider that an app with 100,000+ installs has 100,000 installs, and an app with 1,000,000+ installs has 1,000,000 installs, and so on.

To perform computations, however, we'll need to convert each install number to float — this means that we need to remove the commas and the plus characters, otherwise the conversion will fail and raise an error. We'll do this directly in the loop below, where we also compute the average number of installs for each genre (category).



```python
categories_android = freq_table(android_final, 1)

for category in categories_android:
    total = 0
    len_category = 0
    for app in android_final:
        category_app = app[1]
        if category_app == category:            
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
```
